```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
dataset = pd.read_csv('algerian+forest+fires+dataset//Algerian_forest_fires_dataset_UPDATE.csv',header=1) 
dataset
```

```python
dataset.info()
```

## Data Cleaning

```python
# missing values 
dataset[dataset.isnull().any(axis=1)]
```

```python
# the dataset is coverted into two sets based on the regions:
# 1."Bejaia Region dataset" 
# 2."Sidi-Bel Abbes Region Dataset"
```

```python
dataset.loc[:122,"Region"]=0 
dataset.loc[122:,"Region"]=1  
df =dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# removing the 122th row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fixing spaces in columns names 
df.columns = df.columns.str.strip() 
df.columns
```

```python
# Changing the required columns as integer data type 
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
# similarly changing other columns as float 
df[['Rain', 'FFMC',
'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC',
                                        'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
df.describe() 
```

```python
 # saving the clean dataset
df.to_csv('Algerian_Forest_Fires_Cleaned_Dataset.csv',index=False)
```

## Exploratory Data Analysis

```python
# percentage for pie chart 
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting pie chart 
classlabels = ['Fire','Not Fire'] 
plt.figure(figsize=(12,7)) 
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%') 
plt.show()
```
